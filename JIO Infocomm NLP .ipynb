```python
pd.set_option('display.max_colwidth',100)

df = pd.read_csv(r"C:\Users\Acer\Desktop\Rishi\JIO Internship\jio projects\Jio complaint forum NLP\scrapped_data.tsv", sep="\t",header = None)
df.columns = ["Complaints"]

df
```

```python
df.index = ["Tab 1","Tab 2","Tab 3","Tab 4","Tab 5","Tab 6","Tab 7","Tab 8","Tab 9","Tab 10"]
df.head(3)
```

```python
import string
import re
import nltk
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def clean_text(text):
    punctuation_less_text = "".join([char for char in text if char not in string.punctuation])
    tokens= re.split('\W+', punctuation_less_text)
    stopword_less_tokens = [word for word in tokens if word not in stopwords]
    return(stopword_less_tokens)
```

```python
df["cleaned text"] = df['Complaints'].apply(lambda x : clean_text(x.lower()))
pd.set_option('display.max_colwidth',100)
df.head(3)
```

```python
wn = nltk.WordNetLemmatizer()

def lemmatizing(tokenized_text):
    lemmatized_text = [wn.lemmatize(word) for word in tokenized_text]
    return(lemmatized_text)

df['Lemmatized text'] = df['cleaned text'].apply(lambda x: lemmatizing(x))
df.head(3)
```
